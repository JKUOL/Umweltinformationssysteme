```python
import pandas as pd
import numpy as np
```

```python
# Read CSV files
NO_df = pd.read_csv("city_data/NO2.csv", sep=';')
NO_hist_OL_df = pd.read_csv("city_data_historic/Oldenburg_VS_181228_190328.csv", sep=';')

# Convert 'Datum' column to date format
NO_hist_OL_df['Datum'] = pd.to_datetime(NO_hist_OL_df['Datum'], format="%d.%m.%y")
```

```python
print(NO_hist_OL_df.columns.tolist())
print(NO_df.columns.tolist())
```

```python
# Replace comma with dot and convert to float
NO_df['Oldenburg_VS'] = NO_df['Oldenburg_VS'].str.replace(',', '.').astype(float)
```

```python
# Calculate running mean for NAs
window_size = 48  # Number of elements used for running mean

running_mean = NO_df['Oldenburg'].rolling(window=window_size, min_periods=1).mean()
```

```python
NO_hist_OL_df
```

```python
# Replace missing values with the running mean
NO_hist_OL_df['ug.m'] = NO_hist_OL_df['ug.m'].mask(NO_hist_OL_df['ug.m'].isna(), running_mean)

# Create a new DataFrame with the desired columns
NO_OL_df = pd.DataFrame({'NO_df_Datum': NO_df['Datum'],
                         'NO_df_Uhrzeit': NO_df['Uhrzeit'],
                         'NO_df_Oldenburg_VS': NO_df['Oldenburg_VS']})
```

```python
# Rename columns
NO_OL_df.columns = ['Datum', 'Uhrzeit', 'ug/m^3']

# Convert 'Datum' column to date format
NO_OL_df['Datum'] = pd.to_datetime(NO_OL_df['Datum'], format="%d.%m.%y")

# Calculate running mean for NAs
window_size = 48  # Number of elements used for running mean
running_mean = NO_df['Oldenburg_VS'].rolling(window=window_size, min_periods=1).mean()

# Replace missing values with the running mean
NO_OL_df['ug/m^3'] = NO_OL_df['ug/m^3'].mask(NO_OL_df['ug/m^3'].isna(), running_mean)
```
